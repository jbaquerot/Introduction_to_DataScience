# Titanic: Data Exploration

## Libraries

```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

## Loading Data

```python
passengers = pd.read_csv("data/titanic/train.csv")
```

## Univariate Data

```python
passengers.columns
```

```python
ages = passengers['Age']
ages.describe()
```

```python
passengers.head()
```

```python
plt.figure(figsize=(8,8))
ax = ages.hist(bins = 8, alpha=0.5)
ax.set_xlabel("Age")
ax.set_ylabel("# of passengers")
ax.set_title("Age histogram")
```

## Bivariate Data

```python
ages = passengers['Age']
fares = passengers['Fare']
```

```python
plt.figure(figsize=(8,8))
ax = plt.scatter(fares, ages, alpha=0.2)
plt.title("Scatter plot Fare vs Age")
```

```python
plt.figure(figsize=(8,8))
colors = [['r','g','b'][i-1] for i in passengers['Pclass']]
ax = plt.scatter(passengers['Age'], passengers['Pclass'], c= colors)
plt.title("Scatter plot Fare vs Class")
```
